```python
import time
import requests
import bs4
```

```python
req = requests.get('https://www.procyclingstats.com/race/tour-de-france/2023/startlist/alphabetical')
soup = bs4.BeautifulSoup(req.text, 'html.parser')

table = soup.find('table', {'class': 'basic'})

riders = []
for i in table:
    for j in i.findAll('tr'):
        for k in j.findAll('td'):
            try:
                riders.append('https://www.procyclingstats.com/'+k.a['href'])
            except:
                pass

print(riders)

with open('riders.txt', 'w') as f:
    for i in riders:
        f.write(i+'\n')
```
